```python
#Import files
import pandas as pd
import numpy as np
```

```python
# Load the data
df = pd.read_csv("/content/Sample - Superstore.csv", encoding='ISO-8859-1')
```

```python
#Info
print(df.head())
```

```python
#Check missing values
print(df.isnull().sum())
```

```python
#Remove duplicates
df = df.drop_duplicates()
```

```python
print(df.info())
```

```python
print(df.columns)
```

```python
# Drop 'Row ID' and 'Product Name' columns
df = df.drop(['Row ID', 'Product Name'], axis=1)

# Check the new shape and columns
df.head()
```

```python
# Convert dates if needed
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
# Convert dates if needed
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
#Check state name
df['State'].value_counts()
```

```python
#Check city name
df['City'].value_counts()
```

```python
#Save clean data
df.to_csv("cleaned_data_task_2.csv", index=False)
```
